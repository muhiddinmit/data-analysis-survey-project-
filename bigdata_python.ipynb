```python
missing = df.isnull().sum().sort_values(ascending=False)
missing_percent = (missing / len(df)) * 100

missing_percent.head(30)
```

```python
"""
we need to do :

> 50%	       DROP
10–50%	     FILL
< 10%	       SAFE, its Okay

"""
```

```python
cols_to_drop = missing_percent[missing_percent > 50].index

df = df.drop(columns=cols_to_drop)
```

```python
#Then we need to fill missing values (Categorical)

cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    df[col] = df[col].fillna("Unknown")

```

```python
#Also we need to fill missing values (Numeric)

num_cols = df.select_dtypes(include='number').columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Also we need to fix data tpyes columns like YearsCode,Age are messy

df['YearsCode'] = pd.to_numeric(df['YearsCode'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

```

```python
# Removing Outliers,  it is important for salary

if 'ConvertedCompYearly' in df.columns:
    df = df[df['ConvertedCompYearly'] < df['ConvertedCompYearly'].quantile(0.99)]
```

```python
# Important Columns (for analysis)

cols = [
    'Country',
    'Age',
    'EdLevel',
    'YearsCode',
    'Employment',
    'ConvertedCompYearly'
]

df = df[[col for col in cols if col in df.columns]]
```
